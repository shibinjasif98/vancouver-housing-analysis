# Vancouver Housing Analysis
## Data sources & Research questions

# Data sources
We will use publicly available datasets to analyze housing
affordability in Vancouver

Planned sources:
- Vancouver Open Data Portal (housing and rental-related datasets)
- Statistics Canada (income and economic indicators)
- CMHC rental market data

# Key Reasearch Questions
1. How have rental prices in vancouver changed over time?
2. How does income grouwth compare to rent growth?
3. Are there identifiable affordabiltiy patterns across
   areas (if location data is available)?
4. What do these trends imply about housing affordability
   today?
